## 04 – Escreva o programa em Python:

```python
import numpy as np 

array_dez = np.random.normal(loc = 10, scale = 3, size = 5)
maior_10 = array_dez[array_dez > 10]
tamanho = len(maior_10)
print (array_dez)
print (f"{tamanho} valores são maiores que 10, sendo eles: {maior_10}")
```

## 09 – Escreva o programa em Python:

```python
import pandas as pd 

df = pd.read_csv('genes.csv', index_col=0)
primeiro_gene = df.index[0]
n_char = len(primeiro_gene)

if 2 <= n_char <= 4:
    print("Posso saber as informações do gene:")
    print(df.loc[primeiro_gene])

elif n_char < 2:
    print("Valores onde os nomes têm mais de 4 caracteres:")
    print(df[df.index.str.len() > 4])

else:
    print("Genes com nomes maiores que 4 caracteres:")
    print(df.index[df.index.str.len() > 4])
```
