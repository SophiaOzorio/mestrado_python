## 01 – Dados as operações abaixo, qual o resultado da operação lógica: 
p and q and r and s and t?
<br>
<br>Como os operadores lógicos podem ser usados para inverter o resultado?

<br>p = 2.0 + 3.0 / 5.0 == 10
<br>q = (p + 10.2) - 5.0 * 2.0 <= 2.9
<br>r = 10 <= p / q + 33.8 < 100
<br>s = 5 != p / (q + r)
<br>t = 33 == s * 2

```python
p = 2.0 + 3.0 / 5.0 == 10
q = (p + 10.2) - 5.0 * 2.0 <= 2.9
r = 10 <= p / q + 33.8 < 100
s = 5 != p / (q + r)
t = 33 == s * 2

resultado = (p and q and r and s and t)
inverter = (p or q or r or s or t)
print (f"(p and q and r and s and t) -> {resultado}")

print (f"Para inverter o resultado pode-se usar o operador lógico 'or': ")
print (f"(p or q or r or s or t) -> {inverter}")
```

## 02 – Usando apenas o operador or, qual seria o resultado final das operações abaixo? Como deve ser usado o not para inverter o resultado?

<br>p = ('a',1,True,'b',0,False) != ('a',True,1,'b',False,0)
<br>q = ['a',1,True,'b',0,False] != ('a',True,1,'b',False,0)
<br>r = ['a',1,True,'b',0,False] != ['a',True,1,'b',False,0]
<br>s = ['a',1,True,'b',0,False] > ['a',True,1]
<br>t = ['a',1,True,'b',0,False] == 1
<br>u = ['a',True,1] <= ['a',True,1] < ['a',1,True,'b',0,False]

```python
p_dois = ('a',1,True,'b',0,False) != ('a',True,1,'b',False,0)
q_dois = ['a',1,True,'b',0,False] != ('a',True,1,'b',False,0)
r_dois = ['a',1,True,'b',0,False] != ['a',True,1,'b',False,0]
s_dois = ['a',1,True,'b',0,False] > ['a',True,1]
t_dois = ['a',1,True,'b',0,False] == 1
u_dois = ['a',True,1] <= ['a',True,1] < ['a',1,True,'b',0,False]

resultado_2 = (p_dois or q_dois or r_dois or s_dois or t_dois or u_dois)
resultado_3 = not (p_dois or q_dois or r_dois or s_dois or t_dois or u_dois)
print (f"Utilizando apenas o operador 'or' o resultado das operações abaixo é: {resultado_2}")
print (f"O 'not' deve ser usado dessa forma para inverter o resultado: 'not (p_dois or q_dois or r_dois or s_dois or t_dois or u_dois)' -> {resultado_3}")
```

## 03 – No exercício 3 da aula 8 foi importado um arquivo .csv para um objeto da classe DataFrame pandas.
