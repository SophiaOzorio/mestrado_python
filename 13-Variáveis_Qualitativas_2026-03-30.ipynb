## 03 – Escreva o programa em Python:

```python
df = pd.read_csv('genes.csv')
df
```

```python
df['Tamanho_do_Nome'] = df['Unnamed: 0'].astype(str).str.len()

tabela_frequencia = df['Tamanho_do_Nome'].value_counts().reset_index()
tabela_frequencia.columns = ['Numero_de_Caracteres', 'Frequencia']

print(tabela_frequencia)

tabela_frequencia.plot(
    x='Numero_de_Caracteres', 
    y='Frequencia', 
    kind='bar', 
    legend=False,
    color='skyblue',
    edgecolor='black'
)

# Configurações visuais 
plt.title('Frequência do Número de Caracteres nos Nomes dos Genes')
plt.xlabel('Número de Caracteres')
plt.ylabel('Frequência')
plt.xticks(rotation=0) 

plt.show()
```

R.:

3 - Renomeei as colunas da tabela de frequência.<br>
4 - Utilizei o método `plt.bar` para plotar os dados.<br>

## 06 – Escreva o programa em Python:

```python

df['Tamanho_do_Nome'] = df['Unnamed: 0'].astype(str).str.len()

# 3. Criando a Tabela de Frequência Ordenada
tabela_frequencia = df['Tamanho_do_Nome'].value_counts().sort_index().reset_index()
tabela_frequencia.columns = ['Numero_de_Caracteres', 'Frequencia']
print (tabela_frequencia)

# 4. Gráfico de barras
plt.figure(figsize=(10, 5))
plt.bar(
    tabela_frequencia['Numero_de_Caracteres'].astype(str), # Convertido para str para evitar gaps no eixo X
    tabela_frequencia['Frequencia'], 
    color='skyblue', 
    edgecolor='black'
)

# Configurações visuais
plt.title('Análise de Genes: Frequência por Tamanho (Ordenado)')
plt.xlabel('Número de Caracteres no Nome')
plt.ylabel('Quantidade de Genes')
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```
