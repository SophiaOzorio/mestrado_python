## 1 – Um objeto composto multidimensional (matriz) pode receber  objeto composto unidimensional (vetor)?

```python
import numpy as np

matriz = np.zeros((3, 3)) #criei um objeto multidimensional
vetor = np.array([10, 20, 30]) #criei um objeto unidimensional
matriz[1] = vetor #inseri o objeto unidimensional no index 1 do objeto multidimensional

print (f"Sim! Um objeto composto multidimensional pode recerber um objeto composto unidimensional.")
print (matriz)
```

## 2 – No exercício 9 da última aula, foram instanciadas 5 listas. Instancie um objeto composto multidimensional da classe lista com estes cinco vetores. Recupere a matriz.

```python
genes = ['BRCA1','TP53', 'EGFR', 'CFTR', 'APOE']

gene1 = [
    len(genes[0]),                 
    genes[0][:2],                  
    genes[0].lower().count('a'),  
    "hu" in genes[0].lower()      
]

gene2 = [
    len(genes[1]),                 
    genes[1][:2],                  
    genes[1].lower().count('a'),  
    "hu" in genes[1].lower()      
]

gene3 = [
    len(genes[2]),                 
    genes[2][:2],                  
    genes[2].lower().count('a'),  
    "hu" in genes[2].lower()      
]

gene4 = [
    len(genes[3]),                 
    genes[3][:2],                  
    genes[3].lower().count('a'),  
    "hu" in genes[3].lower()      
]

gene5 = [
    len(genes[4]),                 
    genes[4][:2],                  
    genes[4].lower().count('a'),  
    "hu" in genes[4].lower()      
]

obj_composto = [gene1, gene2, gene3, gene4, gene5]
obj_composto
```

## 7 – Converta o objeto instanciado no exercício 10 da aula anterior para uma DataFrame Pandas. Nesta DataFrame, os index devem ser os nomes dos genes e cada coluna deve ter um nome que indique seu conteúdo. Recupere a DataFrame.

```python
import pandas as pd 
dict_3 = {
    genes[0]: gene1,
    genes[1]: gene2,
    genes[2]: gene3,
    genes[3]: gene4,
    genes[4]: gene5
}

df = pd.DataFrame(dict_3).transpose()
df.columns = ["Tamanho", "Inicio", "Qtd_a", "Tem_hu"]
print(df)
```
