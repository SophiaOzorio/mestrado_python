## 04 – Escreva o programa em Python:

```python
import numpy as np
```

```python
def analisar_valores():
    
    array = np.random.normal(loc=10, scale=3, size=5)
    maiores_que_10 = array[array > 10]
    quantidade = len(maiores_que_10)
    
    print("Array gerado:", array)
    print("Valores > 10:", maiores_que_10)
    print("Quantidade:", quantidade)
    
    return array, maiores_que_10, quantidade

analisar_valores()
```

## 09 – Desenvolva um programa de computador que retorne a divisão de dois objetos simples. Seu programa deve ser capaz de reconhecer a classe destes objetos e o valor, antes de retornar ao usuário.

```python
def dividir_simples(a, b):
    print("Objeto A:")
    print("Valor:", a)
    print("Classe:", type(a))
    
    print("\nObjeto B:")
    print("Valor:", b)
    print("Classe:", type(b))
    
    try:
        resultado = a / b
        print("\nResultado da divisão:", resultado)
        print("Classe do resultado:", type(resultado))
    except Exception as e:
        print("\nErro na divisão:", e)

dividir_simples(10, 30)
```

## 10 – Desenvolva um programa de computador que retorne a divisão de dois objetos compostos. Seu programa deve ser capaz de reconhecer a classe destes objetos e o valor, antes de retornar ao usuário.

```python
def dividir_composto(lista_a, lista_b):
    print("Objeto 1:")
    print("Valor:", lista_a)
    print("Classe:", type(lista_a))
    
    print("\nObjeto 2:")
    print("Valor:", lista_b)
    print("Classe:", type(lista_b))

    resultado = []
    
    try:
        for i in range(len(lista_a)):
            resultado.append(lista_a[i]/lista_b[i])
            
        print("\nResultado da divisão:", resultado)
        print("Classe do resultado:", type(resultado))
    except Exception as e:
        print("\nErro na divisão:", e)

dividir_composto([10, 30, 40, 12], [2, 60,10,5])
```
